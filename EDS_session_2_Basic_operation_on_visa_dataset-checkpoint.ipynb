**Task-1**

- I want to create two lists

- One is categorical column list

- Second one is numerical column list

- Using dtypes

- Whenever you want to create any list from series 

- First convert the series into a dictionary

- When you convert it becomes a key : value pair

- Iterate through dictionary

- If that value == object type, then keep that corresponding key in one list

- else keep that variable in other list

```python
data_types=visa_df.dtypes
dict(data_types)
```

```python
cat_list=[]
num_list=[]
for i,j in dict(data_types).items():
    if j=='object':
        cat_list.append(i)
    else:
        num_list.append(i)

cat_list,num_list
```

```python
cat_list=[i for i,j in dict(data_types).items() if j=='object']
num_list=[i for i,j in dict(data_types).items() if j!='object']
cat_list,num_list
```

- Take dtypes

- Convert into dictionary

- Create 2 lists

- create two dataframes

```python
len(cat_list),len(num_list)
```

- We can not create a dataframe using these two lists

- Because these two ilsts are not in equal length

- If you want to c reat a dataframe, then we need to make two lists has equal length 

    - Then youneed to question yourself 

    - Making equal length required or not 

    - If it is require how ?

**Create columns dataframe**

```python
num_df=pd.DataFrame(num_list,columns=['Numerical_column'])
num_df.to_csv('Numerical_column_df.csv',index=False)
```

```python
cat_df=pd.DataFrame(cat_list,columns=['Categorical_column'])
cat_df.to_csv('Categorical_column_df.csv',index=False)
```

$Take$

- First argument : indices

- Indices will take some values in array : in a list

- It tries to extract the data

- That values we provided represents columns or rows ?

- Axis = 0 represents about rows

- Axis = 1 represents about columns

- By default axis mentioned as zero

```python
visa_df.take([1],axis=1)   # Columns
```

```python
visa_df.take([100,200,300])  

# It retrives 3 rows which having index 100,200,300
```

```python
# Case-2
visa_df.take([100,200,300],axis=1)  

# now this will check for 100th column, 200 column and 300 column
# But originally we have only 12 column (0,11)
# Will get error
```
