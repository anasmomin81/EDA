**LableEncoder**

- LableEncoder is a method to convert categorical data to Numerical data

- LableEncoder is a same way, how we did in the above

- The for loop map method we explored

- But the LableEmcoder also do the same

- It is under **sickit-learn** we usually called as **sklearn**

- Uder skleran we have class called as preprocessing

- Because currently we are doing data preprocessing only

- Under preprocessing class we have a method called as **LableEncoder**

- sklearn

    - Preprocessing

    - LableEncoder
 
- Any sklearn package has 3 steps

    - Step-1: Read the package

    - Step-2: Save the package
 
    - Step-3: Apply fit transform on data

```python
visa_df=pd.read_csv('Visadataset.csv')
# Step-1: Read the package
from sklearn.preprocessing import LabelEncoder
# Step-2: Save the package or call the method
le=LabelEncoder()
# Step-3: Apply fit transform on data
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df
```

```python
visa_df=pd.read_csv('Visadataset.csv')
cat_column=visa_df.select_dtypes(include='object').columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat_column:
    visa_df[i]=le.fit_transform(visa_df[i])
visa_df
```

**fit and transform**

- fit means develop the logic 

- transform means process the logic

- If we observe in map method first we developed dictionary logic

- The we apply that dictionary to column

- Developing the dictionary is kind of fit

- Apply the dictionary to column means we are tranforming the data cat to num

- Whenever we are tranforming or changing the data then use **fit transform**

- If we are not changing the data then only **fit**

**np.where**

- using np.where also we can change the categoricldata to Numerical

- But np.where is only for binary condition

- np.where is same as if-else

- If conditoin is True : will apply True value

- If conditoiin is False: will apply False value

- So np.whee use for only binary lables

- A column should have only two unique lables

```python
# value=='certified  replace with 0
# otherwise 1
# np.where (con,0,1)
```

```python
visa_df=pd.read_csv('Visadataset.csv')
con=visa_df['case_status']=='Certified'
visa_df['case status']=np.where(con,0,1)
visa_df
```

**One hot endorder**

- One hot means if one will be ON, another will be OFF

- ON represents with 1

- OFF represents with 0

- For example case status has two unique lables

    - Certified

    - Denied
 
- one hot encoder creates new columns which is equal to number of unique lables

- For example case status will create two new extra columns

    - case_status_Certified
 
    - case_status_Denied
